```python
# функция для выполнения действий над матрицами из Задачи 1
def operation(matrix1,matrix2):
  if matrix1.shape != matrix2.shape: # сравниваем размеры складываемых матриц
    matrix2 = matrix2.transpose() # транспонируем матрицу 2, т.к. сложить можно только матрицы одинакового размера
  else:
    matrix2 = matrix2
  return matrix1+2*matrix2
```

```python
# A+2*B
A = np.matrix([[1, 2, 1], [-2, 1, 1]])
B = np.matrix(([[3, 1], [-1, 0], [2, 4]]))
print(operation(A,B))
```

```python
# B+2C
B = np.matrix(([[3, 1], [-1, 0], [2, 4]]))
C = np.matrix(([[-2, 2], [1, -1], [1, 2]]))
print(operation(B,C))
```

```python
def сhess():
  D = np.zeros((8,8),dtype=int) # создаем матрицу 8*8, состоящую из нулей
  D[1::2,::2] = 1 # заполняем единицами чётные строки
  D[::2,1::2] = 1 # заполняем единицами нечётные строки
  return D
print(сhess())
```

```python
# Формируем два вектора
f1 = np.arange(start=0, stop=1.5, step=0.1, dtype=float)
f2 = np.arange(start=1, stop=7, step=1, dtype=int)

# преобразовываем вектора в матрицы
f1.shape = (5, 3)
f2.shape = (3, 2)

# перемножаем матрицы
f3=np.dot(f1,f2)

# выводим сами матрицы и результат их перемножения на экран
print(f1)
print(f2)
print(f3)
```
