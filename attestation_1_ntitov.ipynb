```python
dataset_tk = dataset_tk.rename(columns={'Unnamed: 0': 'Date'}) #    
plt.rcParams ['figure.figsize'] = [16, 7] #   
```

```python
def graph(City):
  '''    '''
  dataset_tk.set_index("Date")[City].plot()
```

```python
graph("Punjab")
graph("Haryana")
graph("Rajasthan")
graph("Delhi")
plt.title(' ')
plt.ylabel(' ')
plt.xlabel('')
plt.legend(['Punjab', 'Haryana', 'Rajasthan',	'Delhi'])
plt.show()
```
