```python
import pandas as pd # подключаем библиотеку Pandas
import matplotlib.pyplot as plt # подключаем библиотеку MatPlotLib
import numpy as np # подключаем библиотеку NumPy
```

```python
dataset_tk = pd.read_csv("https://raw.githubusercontent.com/NikitaLipp/AIarchitecture/main/dataset_tk.csv")
```

image.png

```python
# Определяем общее количество пустых ячеек
number_of_empty_cells = dataset_tk.isnull().sum().sum()
print("Количество пустых ячеек:", number_of_empty_cells)
```

```python
# Считаем количество уникальных значений по столбцам
unique_values = dataset_tk.nunique()
print (unique_values)
```

```python
# Выводим минимальные значения по столбцам
min_column = dataset_tk.min(numeric_only=True)
print(min_column)
```

```python
# Выводим максимальные значения по столбцам
max_column = dataset_tk.max(numeric_only=True)
print(max_column)
```

```python
# Рассчитываем отношение макс и мин
max_column / min_column
```

```python
# Выводим максимальные значения по столбцам
mean_column = dataset_tk.mean(numeric_only=True)
print(mean_column)
```

```python
# Выводим общую описательную статистику
dataset_tk.describe()
```

```python
dataset_tk = dataset_tk.rename(columns={'Unnamed: 0': 'Date'}) # Называем корректно первый столбец
plt.rcParams ['figure.figsize'] = [16, 7] # Задаём формат графиков
```

```python
def graph(City):
  '''Функция отрисовки графика по городу'''
  dataset_tk.set_index("Date")[City].plot()
```

```python
graph("Punjab")
graph("Haryana")
graph("Rajasthan")
graph("Delhi")
plt.title('Потребление электроэнергии')
plt.ylabel('Потребляемая мощность')
plt.xlabel('Дата')
plt.legend(['Punjab', 'Haryana', 'Rajasthan',	'Delhi'])
plt.show()
```
